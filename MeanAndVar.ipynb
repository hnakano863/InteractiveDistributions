```julia
using Distributions, Interact, Plots
gr()

function inferance_picker()
    真の平均 = 170
    真の標準偏差 = 20
    真の分布 = Normal(真の平均, 真の標準偏差)
    
    s = input(100; typ = "number", label = "サンプル数")
    btn = button("Resample!")
    n = onchange(s, btn)
    
    サンプル = Interact.@map rand(真の分布, &n)
    標本平均 = Interact.@map mean(&サンプル)
    標本標準偏差 = Interact.@map std(&サンプル)
    予測分布 = Interact.@map Normal(&標本平均, &標本標準偏差)
    
    xs = 110:230
    ys1 = pdf.(Ref(真の分布), xs)
    ys2 = Interact.@map pdf.(Ref(&予測分布), xs)
    
    p1 = Interact.@map plot(xs, [ys1, &ys2], xlims = (110, 230), ylims = (-Inf, 0.025))
    p2 = Interact.@map histogram(&サンプル, bins = 10)
    p = Interact.@map plot(&p1, &p2, layout = (2, 1))
    hbox(p, hbox(n, btn))
end
```

```julia
inferance_picker()
```
